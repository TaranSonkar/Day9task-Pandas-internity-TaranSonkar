**Data Structures in Pandas**

Pandas is an open-source library that uses for working with relational or labeled data both easily and intuitively. It provides various data structures and operations for manipulating numerical data and time series. It offers a tool for cleaning and processes your data. It is the most popular Python library that is used for data analysis. In this article, We are going to learn about Pandas Data structure.

It supports three data structures:

* Series

**Creating a Series**

Series--
It is one dimension labled array which returns and object int a from of list you can say that a series type of list.
It contains single axis index series can only contain a list with an index
series are size inmutable (you can not change there size).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
```

```python
series_data=[10,'a',19462,'ALPHA',0.255,'B',90,'beta','ZZZ',10000]
series_output=pd.Series(series_data)
series_output
```

**Change index Number**

```python
c=pd.Series(series_data,index=[2,3,4,5,6,7,8,9,99,99])
c
```

**Creting Numpy Series**

```python
series=['alpha',58911,'b',0.525]
se=np.array(series)
print(se)
series_out=pd.Series(se)
print(series_out)
```

**Pannel Data Structures**

A panel is a 3D container of data. The term Panel data is derived from econometrics and is partially responsible for the name pandas − pan(el)-da(ta)-s. items − axis 0, each item corresponds to a DataFrame contained inside. major_axis − axis 1, it is the index (rows) of each of the DataFrames.
Data: Data takes various forms like ndarray, se...
Dtype: Data type of each column
Minor_axis: axis=2
Copy: Copy data. Default, false

**Syntax**

pd.Panel(data=None, items=None,major_axis=None,minor_axis=None,copy=Fasle,dtype=None)

```python
f={'a':'apple','b':'ball','c':'cat'}
f=pd.DataFrame(f,index=[1,2,3])
f
```

```python
p={'a':'Ravi','b':'babu','c':'soona'}
p=pd.DataFrame(p,index=[1,2,3])
p
```

```python
import pandas as pd
import panel as pn
```

```python
pan=pd.Panel(data={'item':f,'item2':p},items=['first','second'],major_axis=[2,3,4],minor_axis=['a','b','c'])
```
